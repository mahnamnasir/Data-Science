```python
# importing library pandas and loading the csv file
import pandas as pd
df =pd.read_csv("/content/weather_data (1).csv",parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method='bfill',limit=1)
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

```python
new_df = df.dropna(how="all")
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
